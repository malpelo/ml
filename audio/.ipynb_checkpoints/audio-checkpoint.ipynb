```python
from audio2numpy import open_audio
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal, signal_rate = open_audio("lisergic.mp3")
```

```python
orig = signal[:, 0]
orig_rms = np.sqrt(np.correlate(orig, orig))
noisies = []
for i in range(10):
  noisy = orig + 1/orig_rms*np.random.uniform(size=len(orig))
  noisy_rms = np.sqrt(np.correlate(noisy, noisy))
  noisy_norm = noisy/noisy_rms
  noisies.append(noisy_norm)
orig = orig/orig_rms
```

```python
fig, a = plt.subplots(2)
a[0].plot(orig[:10000])
a[1].plot(noisies[0][:10000])
plt.show()
```

```python
c = sum([np.correlate(noisy, orig[:-10000]) for noisy in noisies])
plt.plot(c)
plt.show()
```
